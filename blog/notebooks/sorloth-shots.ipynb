# Sorloth Shot Analysis - ATM 24/25

```python
import sys
sys.path.append('../..')
from viz.match_data import extract_match_complete
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

matches_df = pd.read_csv('../data/atm_24-25.csv')
print(f"Total partidos: {len(matches_df)}")
matches_df.head()
```

```python
all_sorloth_shots = []

for idx, match in matches_df.iterrows():
    print(f"\n[{idx+1}/{len(matches_df)}] {match['home_team']} vs {match['away_team']}")
    
    try:
        extract_match_complete(
            ws_id=int(match['whoscored_id']),
            us_id=int(match['understat_id']),
            league=match['league'],
            season=match['season'],
            home_team=match['home_team'],
            away_team=match['away_team'],
            match_date=match['date']
        )
        
        events = pd.read_csv('../../viz/data/match_events.csv')
        shots = events[events['type'].str.contains('Shot|Goal', case=False, na=False)].copy()
        sorloth = shots[shots['player'].str.contains('Srloth|Sorloth', case=False, na=False)].copy()
        
        if len(sorloth) > 0:
            is_home = 'Atltico' in match['home_team'] or 'Atletico' in match['home_team']
            sorloth['match_date'] = match['date']
            sorloth['opponent'] = match['away_team'] if is_home else match['home_team']
            sorloth['venue'] = 'Home' if is_home else 'Away'
            sorloth['match_id_ws'] = match['whoscored_id']
            sorloth['match_id_us'] = match['understat_id']
            all_sorloth_shots.append(sorloth)
            print(f"  {len(sorloth)} tiros Sorloth")
        else:
            print(f"  Sin tiros Sorloth")
            
    except Exception as e:
        print(f"  Error: {e}")
        continue

print(f"\nPartidos con tiros Sorloth: {len(all_sorloth_shots)}")
```

```python
if len(all_sorloth_shots) > 0:
    df = pd.concat(all_sorloth_shots, ignore_index=True)
    
    if 'is_goal' not in df.columns:
        df['is_goal'] = df['type'].str.contains('Goal', case=False, na=False)
    
    if 'body_part' not in df.columns:
        df['body_part'] = df['qualifiers'].str.contains('Head', na=False).map({True: 'Head', False: 'Foot'})
    
    if 'xg' in df.columns:
        df['xg'] = pd.to_numeric(df['xg'], errors='coerce').fillna(0.0)
    else:
        df['xg'] = 0.0
    
    df = df.sort_values('match_date').reset_index(drop=True)
    
    output = '../data/sorloth_all_shots_24-25.csv'
    df.to_csv(output, index=False)
    
    print(f"Guardado: {output}")
    display(df[['match_date', 'opponent', 'venue', 'minute', 'type', 'xg', 'is_goal', 'body_part']].head(10))
else:
    print("Sin tiros Sorloth")
```

```python
if len(all_sorloth_shots) > 0:
    print("ESTADISTICAS SORLOTH 24/25")
    print("="*50)
    
    total = len(df)
    goles = df['is_goal'].sum()
    xg = df['xg'].sum()
    
    print(f"\nGENERAL:")
    print(f"  Tiros: {total}")
    print(f"  Goles: {goles}")
    print(f"  xG: {xg:.2f}")
    print(f"  Conversion: {(goles/total*100):.1f}%")
    print(f"  xG diff: {(goles-xg):+.2f}")
    
    print(f"\nTIPO TIRO:")
    print(df.groupby('body_part').agg({'is_goal': 'sum', 'xg': 'sum', 'player': 'count'}).rename(columns={'player': 'shots'}))
    
    print(f"\nHOME vs AWAY:")
    print(df.groupby('venue').agg({'is_goal': 'sum', 'xg': 'sum', 'player': 'count'}).rename(columns={'player': 'shots'}))
    
    print(f"\nTOP PARTIDOS:")
    print(df.groupby(['match_date', 'opponent']).size().sort_values(ascending=False).head(5))
```
