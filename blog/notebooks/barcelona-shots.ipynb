```python
# Load complete match data for season 2025/26
df_2526 = pd.read_csv('../data/barcelona_matches_2526_complete.csv')

print(f'Total events loaded: {len(df_2526)}')
print(f'Date range: {df_2526["match_date"].min()} to {df_2526["match_date"].max()}')
print(f'Teams: {df_2526["team"].unique()}')

# Filter shot events only
shots_2526 = df_2526[df_2526['type'].isin(shot_types)].copy()

print(f'\nTotal shot events: {len(shots_2526)}')
print(f'Shot types: {shots_2526["type"].value_counts().to_dict()}')

# Split Barcelona shots vs opponent shots
barcelona_shots_2526 = shots_2526[shots_2526['team'].str.contains('Barcelona', case=False, na=False)].copy()
opponent_shots_2526 = shots_2526[~shots_2526['team'].str.contains('Barcelona', case=False, na=False)].copy()

print(f'\nBarcelona shots: {len(barcelona_shots_2526)}')
print(f'Barcelona goals: {barcelona_shots_2526["is_goal"].sum()}')
print(f'Barcelona xG: {barcelona_shots_2526["xg"].sum():.2f}')

print(f'\nOpponent shots: {len(opponent_shots_2526)}')
print(f'Opponent goals: {opponent_shots_2526["is_goal"].sum()}')
print(f'Opponent xG: {opponent_shots_2526["xg"].sum():.2f}')

# Export CSVs
barcelona_shots_2526.to_csv('../data/barcelona_shots_2526.csv', index=False)
opponent_shots_2526.to_csv('../data/barcelona_shots_against_2526.csv', index=False)

print('\nExported:')
print('  - barcelona_shots_2526.csv')
print('  - barcelona_shots_against_2526.csv')
```
